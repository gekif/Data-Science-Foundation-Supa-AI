# Mengevaluasi Kualitas Data

```python
import pandas as pd

# Load dataset
data = pd.read_csv('dataset.csv')

# Cek missing values
print('Missing Values:')
print(data.isnull().sum())

# Cek duplikat
print('Duplicate Rows:')
print(data.duplicated().sum())

# Cek statistik dasar
print('Statistics Summary:')
print(data.describe())
```

# Membuat Visualisasi Data Sederhana

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grafik batang penjualan per kategori
sns.barplot(x='Kategori', y='Jumlah Terjual', data=data)
plt.title('Penjualan per Kategori')
plt.show()
```
