# Mengevaluasi Kualitas Data

```python
import pandas as pd

# Load dataset
data = pd.read_csv('dataset.csv')

# Cek missing values
print('Missing Values:')
print(data.isnull().sum())

# Cek duplikat
print('Duplicate Rows:')
print(data.duplicated().sum())

# Cek statistik dasar
print('Statistics Summary:')
print(data.describe())
```

# Membuat Visualisasi Data Sederhana

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grafik batang penjualan per kategori => Berdasarkan rata-rata jumlah terjual
sns.barplot(x='Kategori', y='Jumlah Terjual', data=data)
plt.title('Penjualan per Kategori')
plt.show()
```

Grafik Seaborn Sesuai Ekspektasi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = pd.DataFrame({
    'ID': [1, 2, 3],
    'Nama Produk': ['Sepatu', 'Tas', 'Laptop'],
    'Kategori': ['Fashion', 'Fashion', 'Elektronik'],
    'Harga': [500000, 300000, 7000000],
    'Tanggal Penjualan': ['2025-05-01', '2025-05-02', '2025-05-03'],
    'Jumlah Terjual': [10, 5, 2]
})

# Hitung total penjualan per kategori
total_penjualan = data.groupby('Kategori')['Jumlah Terjual'].sum().reset_index()

# Plot
sns.barplot(x='Kategori', y='Jumlah Terjual', data=total_penjualan)
plt.title('Total Penjualan per Kategori')
plt.show()
```
